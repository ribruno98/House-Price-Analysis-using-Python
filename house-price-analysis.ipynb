# House Price Analysis

Data analysis project exploring house prices using Python, Pandas, Seaborn, and Linear Regression to identify key factors that influence property values.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.linear_model as linear

Housing = pd.read_csv("Housing.csv")
#Housing columns:price, area, bedrooms, bathrooms, stories,	mainroad,	guestroom, basement,	hotwaterheating,	airconditioning,	parking,	prefarea,	furnishingstatus
font1={'family':'serif', 'color':'blue', 'size':20}
font2={'family':'serif', 'color':'red', 'size':18}
font3={'family':'serif', 'color':'green', 'size':18}

sns.histplot(Housing["price"], bins=30, kde=True)
plt.title("Price distribution", fontdict=font1)
plt.xlabel("Price", fontdict=font3)
plt.ylabel("Count", fontdict=font2)
```

```python
sns.regplot(x="area", y="price", data=Housing, line_kws={"color":"red"})
plt.title("Relationship between area and price", fontdict=font1)
plt.xlabel("Area", fontdict=font3)
plt.ylabel("Price", fontdict=font2)
```

```python
sns.boxplot(x="bedrooms", y="price", data=Housing)
plt.title("Price vs Number of Rooms", fontdict=font1)
plt.xlabel("Bedrooms", fontdict=font3)
plt.ylabel("Price", fontdict=font2)
```

```python
sns.boxplot(x="bathrooms", y="price", data=Housing)
plt.title("Price vs. Number of Bathrooms", fontdict=font1)
plt.xlabel("Bathrooms", fontdict=font3)
plt.ylabel("Price", fontdict=font2)
```

```python
sns.heatmap(Housing.select_dtypes(include=["number"]).corr(), annot=True, cmap="coolwarm")
```

Conclusions:



*  The price of houses increases proportionally with the area (positive correlation).
*   Properties with more bedrooms and bathrooms tend to be more expensive.

*   The linear regression model shows that area is the main factor influencing price.
*   Most houses fall within the medium price range.
